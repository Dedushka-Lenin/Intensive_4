```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('date/annotated_data.csv')

columns_of_interest = [' ',
                       '   ',
                       '   ',
                       '   ',
                       '  ']

statistics = {
    ' ':df[' '].sum(),
    '   ':df['   '].sum(),
    '   ':df['   '].sum(),
    '   ':df['   '].sum(),
    '  ':df['  '].sum(),
    '':df[''].sum()
}

#    
labels = list(statistics.keys())
counts = list(statistics.values())

#   
plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color='skyblue')

plt.xticks(rotation=65)

#    
plt.xlabel('')
plt.ylabel('')
plt.title('     ')

plt.tight_layout()
plt.show()
```
