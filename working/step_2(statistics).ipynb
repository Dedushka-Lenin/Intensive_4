```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('date/annotated_data.csv')

columns_of_interest = ['Вопрос решен',
                       'Нравится качество выполнения заявки',
                       'Нравится качество работы сотрудников',
                       'Нравится скорость отработки заявок',
                       'Понравилось выполнение заявки']

statistics = {
    'Вопрос решен':df['Вопрос решен'].sum(),
    'Нравится качество выполнения заявки':df['Нравится качество выполнения заявки'].sum(),
    'Нравится качество работы сотрудников':df['Нравится качество работы сотрудников'].sum(),
    'Нравится скорость отработки заявок':df['Нравится скорость отработки заявок'].sum(),
    'Понравилось выполнение заявки':df['Понравилось выполнение заявки'].sum(),
    'Другое':df['Другое'].sum()
}

# Разделяем ключи и значения
labels = list(statistics.keys())
counts = list(statistics.values())

# Строим столбчатую диаграмму
plt.figure(figsize=(8, 6))
plt.bar(labels, counts, color='skyblue')

plt.xticks(rotation=65)

# Добавляем подписи и заголовок
plt.xlabel('Столбцы')
plt.ylabel('Количество')
plt.title('Распределение единиц и нулей по столбцам')

plt.tight_layout()
plt.show()
```
